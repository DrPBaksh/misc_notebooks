## Indexing Selecting and Filtering
[Back to Contents](#Contents)


```python
import pandas as pd
import numpy as np
obj=pd.Series(np.arange(5),
              index=["a","b","c","d","e"])
```

```python
obj["c"] #  retunrs the record belonging to the index c
obj[2] #  we can also call it by its position
obj[0:3] # select range objects 0, 1, 2
obj[['a', 'c']] # select values by their object name
obj[[0, 2]] # select specific object positions
obj[obj < 2] # return only values who value i less than 2
```

Now lets look at index and filtering in a dataframe

```python
data=pd.DataFrame(
    np.arange(16).reshape(4,4),
    index=["London","Paris",
           "Berlin","London"],
    columns=["one","two","three","four"])
data
```

```python
data['two'] #  Select a column from the dataframe by its name
data.loc['London'] #  get the record by searching using loc for its index
data[['one', 'two']] # select multiple columns
data[:3] # selects the first 3 rows..
data[data > 5] # filter the data
data.query('two < 5') #  filter a column
```

```python
# Selecting rows where loc and iloc. Loc is used when you want to select rows based on the name of their index whereas iloc is used when you select rows based on their record number
```

```python
data.iloc[0] #  record 0
data.iloc[0, 0:2] #  record 0 columns 1 and 2
data.iloc[0, [0, 2, 3]] #  record 0 and specicifcally named columns
```

```python
data.loc['London'] # all london records
data.loc['London', 'one'] #  all london records columns one
data.loc['London', ['one', 'two']]
```

## Combining and Merging
[Back to Contents](#Contents)

```python
d1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
d2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d1,  d2)
```

```python
pd.merge(d1, d2) #  the merge in this case merged on key but only where there was common values. Effectively it was an inner join on key
```

```python
pd.merge(d1, d2, on='key') #  best practise to explicitly state what you are merging on 
```

```python
pd.merge(d1, d2, on='key', how ='left') #  left join
pd.merge(d1, d2, on='key', how ='outer') #  outer join
```

```python
df1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6),
     "count":["one","three","two",
              "one","one","two"]})
df2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4),
     "count":["one","two","two","two"]})
print(df1, df2)
```

```python
pd.merge(df1, df2, on=['key', 'count'], 
         how='outer') # observe here how by requesting a join on both values we are forcing both columns to have identical values as we search for them
```

```python
pd.merge(df1, df2, on=['key', 'count'], 
         how='inner') # this could be clearer on an inner join. Only vales where both key and count were common are merged in the table
```

```python
pd.merge(df1, df2, 
         on='key', 
         suffixes=('_data1', '_data2')) #  where we have column names that are common between table we can add a suffix or prefix to identify which column belonged to which table
```

```python
df1=pd.DataFrame(
    {"letter":["a","a","b",
               "b","a","c"],
     "num":range(6)}) 
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
print(df1, df2)
pd.merge(df1, df2, left_on ='letter', right_index = True) #  Declare that the join will take place for the left hand table on letter and the right hand table on index. 
```

```python
right=pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])
left=pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"])
```

```python
right, left
pd.merge(left, right, left_index = True, right_index = True, how = 'outer')
```

Note on pivot tables

A pivot data turns the data table around so you can look at it from different perspectives. 
In the example below we want to see the amount of fruit that is of that colour. We need to have a table where the index is the fruit and the columns are the colour. We can then see the frequency by thos values as the pieces


```python
# pivot this stock table
stock.pivot(columns = 'fruit') # pivoting makes the target column the column header
stock.pivot(index = 'fruit', columns = 'color') #  see the unique values of 
# stock.pivot()
```

```python
# This function shall help us plot some offset sine waves:
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```
